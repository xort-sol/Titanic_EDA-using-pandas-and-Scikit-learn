Data Cleaning:
Handle missing values in the dataset by performing appropriate imputation techniques. You can consider imputing missing values for 'Age', 'Fare', and 'Embarked' columns.


```python

mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)

median_fare = df['Fare'].median()
df['Fare'].fillna(median_fare, inplace=True)

df['Cabin'].fillna('Unknown', inplace=True)
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)
```

```python
df.head(20)
```

Create a new feature called 'FamilySize' by adding the 'SibSp' (number of siblings/spouses) and 'Parch' (number of parents/children) columns. This feature represents the total number of family members aboard for each passenger.

```python
df['FamilySize']=df['SibSp']+df['Parch']
df.head(20)
```

•	Title Extraction:
o	Extract the titles (e.g., Mr., Mrs., Miss, etc.) from the 'Name' column for each passenger.
o	Create a new feature called 'Title' containing the extracted titles.


```python
def extract_title(name):
    import re
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return ""

df['Title'] = df['Name'].apply(extract_title)
df.head(20)
```

•	Age Group:
o	Group passengers into different age groups (e.g., child, adult, senior) based on their age.
o	Create a new feature called 'AgeGroup' to represent the age group of each passenger.


```python
age_bins = [0, 18, 65, float('inf')]
age_labels = ['Child', 'Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
df.head(20)
```

•	Fare per Person:
o	Calculate the fare per person by dividing the 'Fare' column by the 'FamilySize' for each passenger.
o	Create a new feature called 'FarePerPerson' to represent 


```python
df['farePerPerson'] = df['Fare'] / df['FamilySize']
df.head(10)
```

•	Label Encoding:
o	Encode categorical features such as 'Sex' and 'Embarked' using label encoding technique to convert them into numerical format suitable for machine learning algorithms.


```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Sex_encoded'] = label_encoder.fit_transform(df['Sex'])
df['Embarked_encoded'] = label_encoder.fit_transform(df['Embarked'].astype(str))
```
